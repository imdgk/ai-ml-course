# Answer to the questions

```python
# 1. Calculate the mean of each of the columns of mtcars.

mtcars.mean()
```

```python
# 2. Get the mean of each row (car) and store it in variable 'mtcarsMeanCarWise'.

mtcarsMeanCarWise = mtcars.mean(axis=1)
```

```python
# 3. Get the median of each column and store it in variable 'medianCarsFeatures'.

medianCarsFeatures = mtcars.median()
```

```python
# 4. Plot the histograms for each of the columns in mtcars using hist() function with bin-size as 10.

mtcars.hist(bins=10, figsize=(20,15))
plt.show()
```

```python
# 5. Plot the density curves for each column of mtcars using 'for' loop and plot command.

counter = 1
for column in mtcars:
    if column != 'name':
        mtcars[column].plot(kind='density', figsize=(4,3), title=str(counter)+') '+column)
        plt.show()
        counter += 1
```

```python
# 6. Plot the correlation for the following attributes: mpg, disp, hp, wt, qsec. 
#    You can use the scatter_matrix command present in pandas.plotting library.

from pandas.plotting import scatter_matrix

attributes = ['mpg', 'disp', 'hp', 'wt', 'qsec']

scatter_matrix(mtcars[attributes], figsize=(15,10))
plt.show()
```

```python
# 7. We use skewness as a measure of symmetry. If the skewness of is zero then the distribution represented is perfectly
#    symmetric. If the skewness is negative, then the distribution is skewed to the left, 
#    while if the skew is positive then the distribution is skewed to the right. 
#    Hence, it can also be considered as a measure of normality since normal distributions are symmetric and hence have 
#    a zero skew. Calculate the skeweness of each column of 'mtcars' using mtcars.skew()

mtcars.skew().sort_values(ascending=False)
```

```python
# 8. Use describe and quantile to calculate the Inter-Quantile spread of 'mpg' attribute of mtcars.

mpg = mtcars['mpg'].copy()

Q1 = mpg.quantile(q=0.25)
Q2 = mpg.quantile(q=0.50)
Q3 = mpg.quantile(q=0.75)
IQR = Q3 -Q1

print('First Quartile (Q1):', Q1)
print('Second Quartile (Q2) / Median:', Q2)
print('Third Quartile (Q3):', Q3)
print('Interquartile Range (IQR=Q3-Q1):', IQR)

mpg.describe()
```

```python
# 9. Plot the box plot for 'mpg' column of mtcars.

mpg.plot(kind='box') # Vertical Boxplot
plt.show()

plt.boxplot(mpg, vert=False, showfliers=True) # Horizontal Boxplot
plt.show()
```

```python
# 10. Calculate the mean, standard deviation and variance of the 'mpg' column

mpg_mean = mpg.mean()
mpg_std = mpg.std()
mpg_var = mpg_std ** 2

print('Mean:', mpg_mean)
print('Standard Deviation', mpg_std)
print('Variance:', mpg_var)
```

```python
# 11. Extract cars with automatic transmission and cars with manual transmission into separate variables 'mtCarsAutomatic' 
#     and 'mtCarsManual' respectively.

# am - transmission (0 = automatic, 1 = manual)

mtCarsAutomatic = mtcars[mtcars['am'] == 0]
mtCarsManual = mtcars[mtcars['am'] == 1]
```

```python
# 12. Plot the box plot for 'mpg' (miles per gallons) for each of the subsets of cars grouped by transmission 
#     (automatic/ manual).

mtCarsAutomatic['mpg'].plot(kind='box', title="MPG for Automatic Cars")
plt.show()

mtCarsManual['mpg'].plot(kind='box', title="MPG for Manual Cars") 
plt.show()
```
